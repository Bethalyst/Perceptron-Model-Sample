```python
# Create a dictionary to map the old values to the new values
mapping = {'no': 0, 'yes': 1}

# Apply the replacement to your column
df['family_history_with_overweight'] = df['family_history_with_overweight'].replace(mapping)
df['FAVC'] = df['FAVC'].replace(mapping)
df['SMOKE'] = df['SMOKE'].replace(mapping)
```

```python
correlate_how = df[['Weight','family_history_with_overweight', 'FAVC', 'FCVC', 'NCP', 'FAF', 'TUE']].corr()
```

```python
sns.heatmap(correlate_how)
plt.show()
```

```python
always_drinker = df[(df['CALC'] == 'Always')]

print(f"Average Age {always_drinker['Age'].mean()}")
print(always_drinker.count())
```

```python
overweight_levels =  ['Overweight_Level_I', 'Overweight_Level_II', 'Overweight_Level_III']
overweight_and_drinks = df[(df['CALC'] == 'Frequently') & (df['NObeyesdad'].isin(overweight_levels))]

print(f"Average Age {overweight_and_drinks['Age'].mean()}")
print(overweight_and_drinks.count())
```

```python
non_smokers = df[(df['SMOKE'] == 0) & (df['NObeyesdad'].isin(overweight_levels))]
print(f"Average Age {non_smokers['Age'].mean()}")
print(non_smokers.count())
print(non_smokers.head())
```

```python
df.sort_values(by =['SMOKE']).head(10)
```

```python
import matplotlib.patches as mpatches 

x_data = non_smokers.Age
y_data = non_smokers.Weight

plt.scatter(
    x_data, 
    y_data, 
    c=non_smokers['Gender'].map({'Male': 'blue', 'Female': 'black'}), 
    alpha=0.5
)

plt.scatter(
    overweight_and_drinks.Age, 
    overweight_and_drinks.Weight, 
    c=overweight_and_drinks['Gender'].map({'Male': 'purple', 'Female': 'yellow'}), 
    alpha=0.5
)

plt.title("Age vs. Weight for Overweight (Kg) Non-smoking and Frequent Drinkers")


male_non_smoker = mpatches.Patch(color='blue', label='Male Non-Smoker')
female_non_smoker = mpatches.Patch(color='black', label='Female Non-Smoker')
male_drinker = mpatches.Patch(color='purple', label='Male Drinker')
female_drinker = mpatches.Patch(color='yellow', label='Female Drinker')

plt.legend(
    handles=[male_non_smoker, female_non_smoker, male_drinker, female_drinker],
    title="Groups"
)

plt.show()
```
