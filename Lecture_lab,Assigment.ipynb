## Understanding How to create a perceptron model

This is a machine Learning model that learn the relationship between inputs and output and suggest weight and bias base on the data given.
To use the perceptron function you'll first import it from the sklearn.linear_model which is part of the scikit-learn library, a Python package for machine learning.


```python
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import MinMaxScaler
```

As part of the sklearn packacge it also contains MinMaxScaler which  normalizes your data so that all features fall within a fixed range, (usually 0 to 1) as it will show below. 

After that, select what and where your data will come from . You can see the attached CSV file of my dataset, which I downloaded to my desktop from [kaggle](https://www.kaggle.com/datasets/abirhasan1703100/single-layer-perceptron-datasetsmall)

```python
df = pd.read_csv('OneDrive/Desktop/Python_Projects/Single Layer Perceptron Dataset.csv')
df
```

**The entire row 13 is null, so i removed that row.**

```python
df = df.drop([13], axis = 0)
df
```

**MinMaxScaler** the hardest for me to understand. So i broke it down to remenber, as i said before it suppose to normalizes your data so that all features fall within a fixed range, (usually 0 to 1) right. But how?, 
*scaler* (just a variable name)*= MinMaxScaler()*  analyze what the smallest value and what the largest value in the column, if want to give a specific parameter you use 
**feature_range** . If you don’t write anything inside the parentheses, it automatically uses (0, 1) by default.  like showed in the code.
\
Next **pd.DataFrame()** creates a data frame object, basically a a table you can work with. within that table **fit_tranform()** basically say "hey the scale you were train in called Scaler, now that you know it , do it for the whole data **(df)** within this new datafame". 

**PS** dont Scale your target columns. because 

```python
scaler = MinMaxScaler(feature_range=(0, 1))  # Default is (0, 1)
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df
```

```python
perceptron = Perceptron()
```

 &uarr; **Creates a new Perceptron model** to train

```python
x = df.iloc[:, :-1]
x
```

```python
y = df.iloc[:, -1]
y
```

Remenber fit() used for the MinMaxScaler it does the same here, except here it not attach to the transform. So this time it only train the perceptron model. where  x → features (inputs)
y → target (labels) where the model  learns some numbers called weights

```python
perceptron.fit(x,y) #Learn from this data.
```

```python
 perceptron.coef_  # a NumPy array that stores the learn weights
```

```python
print("Feature Weights: ")
for feature, weight in zip(df.columns[:-1], perceptron.coef_[0]):
    print(f'{feature}: {weight:.2f}')
```

```python
perceptron.intercept_ # a NumPy array that stores the learn bias
```

```python
print(f'Intercept/bias: {perceptron.intercept_[0]:.2f}')
```
